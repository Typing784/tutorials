这里的问题是，使用生成器类的话，只能使用一次，迭代完一次就没有了，所以可以写成一个函数。

```python
def gentest():
  return (alist[i:i+10] for i in range(0, len(alist), 10))

gen = gentest()
print(list(gen))
gen = gentest()
print(list(gen))
```

生成器（Generator）是一种特殊的迭代器，可以通过函数来创建。生成器可以逐个地生成值，而不是一次性产生所有值，这样可以节省内存空间，并且在某些情况下提高效率。在 Python 中，生成器常用于迭代大型数据集或者在循环中生成值的情况下。1.使用生成器函数定义生成器.2.使用生成器表达式.3.使用内置函数，如range()，map()，filter()等。
因此，可以通过定义一个包含 yield 关键字的函数来创建生成器。每次调用生成器的 __next__() 方法时，函数将从 yield 语句的位置开始执行，直到遇到下一个 yield 语句或者函数结束为止。

```python
def gentest():
  for i in range(0, len(alist), 10):
    yield alist[i:i+10]

gen = gentest()
print(list(gen))
gen = gentest()
print(list(gen))
```

```python
def get_training_corpus():
    return (
        raw_datasets["train"][i : i + 1000]["whole_func_string"]
        for i in range(0, len(raw_datasets["train"]), 1000)
    )


training_corpus = get_training_corpus()
```

```python
def get_training_corpus():
    dataset = raw_datasets["train"]
    for start_idx in range(0, len(dataset), 1000):
        samples = dataset[start_idx : start_idx + 1000]
        yield samples["whole_func_string"]
```

```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
example = '''def add_numbers(a, b):
    """Add the two numbers `a` and `b`."""
    return a + b'''

tokens = old_tokenizer.tokenize(example)
tokens
```

```python
tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, 52000)
```

```python
tokens = tokenizer.tokenize(example)
tokens
```

```python
print(len(tokens))
print(len(old_tokenizer.tokenize(example)))
```
